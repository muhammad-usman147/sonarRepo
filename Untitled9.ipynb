```python
def functional_model():
  inputs = keras.Input(shape=(60,))
  layer1 = Dense(60, activation='relu')(inputs)
  layer2 = Dense(60, activation='relu')(layer1)
  outputs = Dense(1, activation='sigmoid')(layer2)

  model = keras.Model(inputs, outputs)
    
 # Compile model, write code below
  model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
  return model
estimators = []
estimators.append(('standarize' , StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn = functional_model , epochs = 150 , batch_size = 50 )))
pipeline = Pipeline(estimators)
kfold= StratifiedKFold(n_splits = 5 , shuffle = True , random_state= seed)
results = cross_val_score(pipeline , X ,enc_y , cv= kfold)
```

```python

class SubclassModel(tf.keras.Model):
  def __init__(self):
    super(SubclassModel, self).__init__()
    self.dense1= tf.keras.layers.Dense(60, activation= tf.nn.relu)
    self.dense2= tf.keras.layers.Dense(30, activation= tf.nn.relu)
    self.dense3= tf.keras.layers.Dense(1, activation= tf.nn.sigmoid)
  def call(self,inputs):
    x = self.dense1(inputs)
    x= self.dense2(x)
    return self.dense3(x)
def finalModel():
  model = SubclassModel()
  model.compile(loss= 'binary_crossentropy' , optimizer = 'adam' , metrics= ['acc'])
  return model
estimators =[]
estimators.append(('standarize' , StandardScaler()))
estimators.append(('mlp' , KerasClassifier(build_fn= finalModel , epochs = 100 , batch_size= 5 )))
pipeline= Pipeline(estimators)
kfold= StratifiedKFold(n_splits = 10 , shuffle= True , random_state= seed)
results = cross_val_score(pipeline , X , enc_y, cv=kfold) 
print('Subclass larger Model : %2f%%(%2f%%)'% (results.mean()*100, results.std()*100))
```

```python
def build_model():
  model= Sequential()
  model.add(Dense(30, activation='relu' , input_shape=(60,)))

  model.add(Dense(1 , activation='sigmoid'))
  model.compile(optimizer='Adam' , loss= 'binary_crossentropy' , metrics=['acc'])
  return model
```

```python
xmodel = build_model()
```

```python
import math as m
```

```python
train_data = X[:m.floor(len(X)*.70)] #gettting 70% for train
```

```python
test_data = X[m.floor(len(X)*.70):] #rest is the test
```

```python
train_labels = enc_y[:m.floor(len(X)*.70)]
test_labels = enc_y[m.floor(len(X)*.70):] 
```

```python
train_data.shape,test_data.shape,train_labels.shape,test_labels.shape
```

```python
k=4
num_val_samples= len(train_data)//k
num_epochs= 50

for i in range(k):
  print('processing fold#' , i)
  val_data = train_data[i* num_val_samples: (i+1) * num_val_samples]
  val_labels = train_labels[i* num_val_samples : (i+1) * num_val_samples]
  partial_train_data= np.concatenate([train_data[:i* num_val_samples], train_data[(i+1)*num_val_samples:]], axis=0)
  partial_train_labels = np.concatenate([train_labels[:i*num_val_samples], train_labels[(i+1)*num_val_samples:]], axis=0)
  
  model= build_model()
  xmodel.fit(partial_train_data, partial_train_labels , epochs= num_epochs , batch_size=50 , verbose= 0)
  loss , acc = xmodel.evaluate(val_data , val_labels, verbose=0)
  print(acc)
```
